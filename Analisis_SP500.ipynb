```python
data = pd.read_excel('datos usa.xlsx')
data = data.rename(columns = {'Return S&P 500':'Stocks Return', 'Return on bond': 'Corporate Bond Return'})
data = data.append(pd.Series([2022, -0.2, 0, -0.18, 0.08], index=['Year','Stocks Return','T.Bond 10Y rate','Corporate Bond Return','Inflation Rate']), ignore_index=True)
print(data.tail())
```

```python
data['Real Stocks Return'] = ((1+data['Stocks Return']) / (1+data['Inflation Rate']) -1)
data['Real Corporate Bond Return'] = ((1+data['Corporate Bond Return']) / (1+data['Inflation Rate']) -1)
data
```

```python
data['Cartera Clasica 60/40'] = (data['Stocks Return']*0.60 + data['Corporate Bond Return']*0.40)
data['Cartera Conservadora 20/80'] = (data['Stocks Return']*0.20 + data['Corporate Bond Return']*0.80)
data['Cartera Agresiva 80/20'] = (data['Stocks Return']*0.80 + data['Corporate Bond Return']*0.20)
data
```

```python
data['Real Cartera Clasica'] = ((1+data['Cartera Clasica 60/40']) / (1+data['Inflation Rate']) -1)
data['Real Cartera Conservadora'] = ((1+data['Cartera Conservadora 20/80']) / (1+data['Inflation Rate']) -1)
data['Real Cartera Agresiva'] = ((1+data['Cartera Agresiva 80/20']) / (1+data['Inflation Rate']) -1)
data
```

```python
print('CARTERA CLASICA')
print("El rendimiento real medio de la Cartera Clasica 60/40 fue de:", data['Real Cartera Clasica'].mean().round(4))
print("La desviación estandar de la Cartera Clasica 60/40 fue de:", data['Real Cartera Clasica'].std().round(4))
print('\n')
print('CARTERA AGRESIVA')
print("El rendimiento real medio de la Cartera Agresiva 80/20 fue de:", data['Real Cartera Agresiva'].mean().round(4))
print("La desviación estandar de la Cartera Agresiva 80/20 fue de:", data['Real Cartera Agresiva'].std().round(4))
print('\n')
print('CARTERA CONSERVADORA')
print("El rendimiento real medio de la Cartera Conservadora 20/80 fue de:", data['Real Cartera Conservadora'].mean().round(4))
print("La desviación estandar de la Cartera Conservadora 20/80 fue de:", data['Real Cartera Conservadora'].std().round(4))
print('\n')
print('SOLO ACCIONES')
print("El rendimiento real medio de las acciones fue de:", data['Real Stocks Return'].mean().round(4))
print("La desviación estandar de las acciones fue de:", data['Real Stocks Return'].std().round(4))
print('\n')
print('SOLO BONO CORPORATIVOS')
print("El rendimiento real medio de los bonos corporativos fue de:", data['Real Corporate Bond Return'].mean().round(4))
print("La desviación estandar de los bonos corporativos fue de:", data['Real Corporate Bond Return'].std().round(4))
```
